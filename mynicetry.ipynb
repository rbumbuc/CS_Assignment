```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# creating df from csv 

df = pd.read_csv('istherecorrelation.csv', sep=';')

# replacing ',' with '.'

df['WO [x1000]'] = df['WO [x1000]'].str.replace(',','.')
```

```python
# floating column

df['WO [x1000]'] = df['WO [x1000]'].apply(lambda x: float(x))

# x axis plot(years) and y axis(WO[x1000] & NL Beer consumption [x1000 hectoliter])
fig, ax = plt.subplots()
ax.plot(df['Year'], df['WO [x1000]'], color="blue", marker="o")

#labelling x, y1,y2
ax.set_xlabel("Year", fontsize=12)
ax.set_ylabel("WO [x1000]", color="blue",fontsize=12)

#twinn x value for both y axis
ax2=ax.twinx()

ax2.plot(df['Year'], df['NL Beer consumption [x1000 hectoliter]'] ,color="red", marker="o")
ax2.set_ylabel("NL Beer consumption [x1000 hectoliter]", color="red", fontsize=12)
plt.title('NL Beer consumption')

#render the plot
plt.show()

#save plot as a file

fig.savefig('plot.png',
           format='png',
           dpi=300,
           bbox_inches='tight')
```
